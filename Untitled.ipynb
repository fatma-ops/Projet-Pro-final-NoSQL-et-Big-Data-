```python
dataAirlines = pd.read_csv("airlines.csv")
```

```python
df = pd.merge(df, dataAirlines[['IATA_CODE', 'AIRLINE']], left_on='AIRLINE', right_on='IATA_CODE', how='left')

df.drop(columns=['AIRLINE_x'], inplace=True)
df.drop(['IATA_CODE'], axis=1, inplace=True)
df.rename(columns={'AIRLINE_y': 'AIRLINE'}, inplace=True)
df
```

```python
batch_size = 30000

# Découper et insérer par lot
for i in range(0, len(df), batch_size):
    batch = df.iloc[i:i + batch_size].to_dict('records')
    try:
        cursor.insert_many(batch)
    except Exception as e:
        print(f"Erreur lors de l'insertion du lot {i//batch_size}: {e}")

# Fermer la connexion
client.close()
```
